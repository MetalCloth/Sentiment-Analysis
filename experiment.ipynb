```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['Sudoku is a fun and challenging puzzle game','Machine learning is transforming many industries','Golden retrievers are friendly and intelligent dogs','Python is a popular language for AI development','Google Cloud offers powerful APIs for developers','Set theory is fundamental in mathematics','Venn diagrams help visualize logical relationships','TensorFlow is widely used for deep learning']
```

```python
sent
```

```python
### vocab size
vocab_size=10000
one_hot_rep=[one_hot(i,vocab_size) for i in sent]
one_hot_rep
```

```python
### word embedding
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
max_len=10
embed=pad_sequences(one_hot_rep,padding='pre',maxlen=max_len)
embed 
embed.shape
```

```python
dim=10
model=Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=dim, input_length=max_len))
model.compile('adam','mse')
```

```python
model.summary()# Check if the embedding layer is working
```

```python
model.predict(embed)
```
